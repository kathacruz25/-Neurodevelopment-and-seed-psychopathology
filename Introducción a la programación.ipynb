# Introducción a la Programación

###### ¿Qué es un lenguaje de programación?

Es un lenguaje formal que, mediante una serie de instrucciones, le permite a un programador escribir un conjunto de órdenes, acciones consecutivas, datos y algoritmos para crear programas que controlen el comportamiento físico y lógico de una máquina.

Dicho de otra forma, es un sistema estructurado de comunicación, el cual está conformado por conjuntos de símbolos, palabras claves, reglas semánticas y sintácticas que permiten el entendimiento entre un programador y una máquina.

#### TIpos de Lenguajes de Programación

$1.$ **El Lenguaje Máquina:** es el lenguaje de programación que entiende directamente la máquina (computadora). Este lenguaje de programación utiliza el alfabeto binario, es decir, el 0 y el 1.

$2.$ **Lenguajes de programación de bajo nivel:** Son mucho más fáciles de utilizar que el lenguaje máquina, pero dependen mucho de la máquina, es un lenguaje primitivo.

$3.$ **Lenguajes de programación de alto nivel.** Los lenguajes de programación de alto nivel son más fáciles de aprender porque se usan palabras o comandos del lenguaje natural, generalmente del inglés. Este es el caso del BASIC, el lenguaje de programación más conocido.

Tipos de lenguajes de programación de alto nivel:

$\cdot$ Lenguaje  imperativo: entre ellos tenemos el Cobol, Pascal, C y Ada.

$\cdot$ Lenguaje declarativo: el Lisp y el Prolog.

$\cdot$ Lenguaje de programación orientado a objetos: el Smalltalk y el C++.

$\cdot$ Lenguaje orientado al problema: son aquellos lenguajes específicos para gestión.

$\cdot$ Lenguaje de programación natural: son los nuevos lenguajes que pretender aproximar el diseño y la construcción de programas al lenguaje de las personas.

# ¿Qué es un Algoritmo?

Pasos $\rightarrow $ verbos $\rightarrow$ obtener un resultado

Instrucciones o pasos (Input-Output)

Programación es el arte de pasar por medio de un lenguaje a la solución de problemas en la vida cotidiana. **Creatividad** pueden existir millones de soluciones para un problema, la diferencia fundamental esta en las intrucciones

**Input** variables (inicio-materiales) cosas exactas que necesito

**Output** resultado final obtenido

**Ejemplo:**  Tomar el Bus

**Input:** 

Parada de autobus (paradero)

**Algoritmo:** 

1) Primero ver el tablero
2) sacar la tarjeta o la plata
3) extiendes tu mano y paras el bus
4) el bus para 

**Output:** 

tomar el bus

# Tipo de datos

Un **array**, es un tipo de dato estructurado que permite almacenar un conjunto de datos homogeneo, es decir, todos ellos del mismo tipo y relacionados.

La clase **ArrayList**, es una clase que permite almacenar datos en memoria de forma similar a los Arrays, con la ventaja de que el numero de elementos que almacena, lo hace de forma dinámica, es decir, que no es necesario declarar su tamaño como pasa con los Arrays.
